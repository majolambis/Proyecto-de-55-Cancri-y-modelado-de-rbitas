# Punto 1
En principio, un asteroide se considera *cercano a la Tierra* (o *Near Earth Asteroid*, NEA) si su órbita puede cruzar la órbita de la Tierra. Matemáticamente, esto se traduce en que la órbita del asteroide debe tener alguna intersección con la órbita terrestre, lo que implica que la distancia mínima del asteroide al Sol (periapsis) sea menor que el radio de la órbita terrestre y que la distancia máxima (apoapsis) sea mayor que el radio de la órbita terrestre.

El asteroide orbita alrededor del Sol en una órbita elíptica con:
- $ q $ la distancia del periapsis del asteroide al Sol.
- $ Q $ la distancia del apoapsis del asteroide al Sol.

La Tierra orbita alrededor del Sol en una órbita elíptica con:
- $ q_\oplus $ la distancia del periapsis de la Tierra al Sol.
- $ Q_\oplus $ la distancia del apoapsis de la Tierra al Sol.

Para la Tierra, dado que su órbita es aproximadamente circular con un semieje mayor de $ a_\oplus = 1 $ UA y una excentricidad pequeña $ e_\oplus = 0.017 $, sus valores de periapsis y apoapsis son aproximadamente:
$$
q_\oplus = a_\oplus (1 - e_\oplus) \approx 0.983 \text{ UA}, \quad Q_\oplus = a_\oplus (1 + e_\oplus) \approx 1.017 \text{ UA}
$$
es decir, que los asteroides con $q < 1.017 UA$ y $Q > 0.989 UA$ cruzan la órbita de la Tierra. Esto lo podemos observar en la imagen:

image.png
Dada una órbita planetaria con un semieje mayor $a$ y una excentricidad $e$
todas las órbitas que tienen $q≤a(1+e)$ y $Q≥a(1-e)$ intersecan la órbita de la Tierra en al menos un punto.

* Si el cuerpo pequeño en su punto más lejano (afelio,Q) llega hasta el punto más cercano de la órbita del planeta o aún más lejos (órbita de color negro).

* Si el cuerpo pequeño en su punto más cercano (perihelio, q) llega hasta el punto más lejano de la órbita del planeta o menos (órbita de color morado).

Recordando que el periapsis $q$ viene de las relaciones en la cónica, es decir, es la distancia desde uno de los focos de la elipse, hasta el punto más cercano. En otras palabras, es la distancia mínima entre dos cuerpos que giran uno alrededor del otro.
Para el caso específico de la Tierra orbitando alrededor del Sol, el periapsis se denomina perihelio.

Por otro lado, el apoapsis $Q$ es el punto más alejado de la órbita al cuerpo centra. En una cónica, este punto representa la distancia máxima entre los dos cuerpos que orbitan.

En el caso de la Tierra orbitando alrededor del Sol, el apoapsis se denomina afelio.

* Las órbitas de los cuerpos individuales son cónicas con la misma excentricidad pero que tienen foco localizado en el centro de masa, en este caso el Sol.



# Condiciones para la intersección de órbitas

Para un cuerpo en órbita alrededor del Sol con semieje mayor \( a \) y excentricidad \( e \), las distancias al perihelio \( q \) y al afelio \( Q \) están dadas por:  

$
q = a(1 - e)
$

$
Q = a(1 + e)
$

Si un asteroide con perihelio $ q_{\text{ast}} $ y afelio $ Q_{\text{ast}} $ debe cruzar la órbita de la Tierra, significa que debe haber al menos un punto en común entre ambas órbitas.  

Para que la órbita del asteroide cruce la órbita de la Tierra, debe cumplirse que:

* Su perihelio sea menor o igual que el afelio de la Tierra:  

   $
   q_{\text{ast}} \leq Q_{\oplus}
   $

   Dado que la órbita terrestre tiene semieje mayor $ a_{\oplus} = 1$ AU y excentricidad $ e_{\oplus} $, su afelio es:

   $
   Q_{\oplus} = a_{\oplus}(1 + e_{\oplus})
   $

   Por lo que la primera condición es:

   $$
   q_{\text{ast}} = a_{\text{ast}}(1 - e_{\text{ast}}) \leq a_{\oplus}(1 + e_{\oplus})
   $$

* Su afelio sea mayor o igual que el perihelio de la Tierra:

   $$
   Q_{\text{ast}} \geq q_{\oplus}
   $$

   El perihelio de la Tierra es:

   $$
   q_{\oplus} = a_{\oplus}(1 - e_{\oplus})
   $$

   Por lo que la segunda condición es:

   $$
   Q_{\text{ast}} = a_{\text{ast}}(1 + e_{\text{ast}}) \geq a_{\oplus}(1 - e_{\oplus})
   $$


* Un asteroide cruzará la órbita de la Tierra si cumple simultáneamente:

$$
q_{\text{ast}} < Q_{\oplus} = a_{\oplus}(1 + e_{\oplus})
$$

$$
Q_{\text{ast}} > q_{\oplus} = a_{\oplus}(1 - e_{\oplus})
$$



En concreto, para que la órbita del asteroide pueda cruzar la órbita de la Tierra, debe existir un punto en la órbita del asteroide donde la distancia al Sol esté dentro del intervalo de la órbita terrestre. Esto implica dos condiciones:

1. **El periapsis del asteroide debe ser menor que el apoapsis de la Tierra**
   $
   q < Q_\oplus
   $:
   Esta condición asegura que la órbita del asteroide puede acercarse lo suficiente al Sol para estar dentro del radio máximo de la órbita terrestre, permitiendo una posible intersección. Es decir, el asteroide puede estar más cerca del Sol que la Tierra en algún punto de su órbita.

2. **El apoapsis del asteroide debe ser mayor que el periapsis de la Tierra**
   $
   Q > q_\oplus
   $:
   Esto garantiza que la órbita del asteroide se extiende más allá del punto más cercano de la órbita terrestre, permitiendo también una posible intersección. Es decir, el asteroide puede estar más lejos del Sol que la Tierra en algún punto de su órbita.

Dado que estos dos criterios aseguran que en algún punto la órbita del asteroide se superpone con la órbita de la Tierra, concluimos que un asteroide es considerado un *Near Earth Asteroid* si y solo si cumple:
$$
q < Q_\oplus \quad \text{y} \quad Q > q_\oplus.
$$

Este resultado nos permite clasificar de manera efectiva los asteroides que tienen el potencial de acercarse a la Tierra.

# Punto 2
* Se toman solo los valores par excentricidades positivas.
* Los cuerpos que pasan la órbita de la Tierra son solo los de la zona sombreda. La zona central del gráfico representa las órbitas que cumplen las condiciones necesarias para cruzar la órbita de la Tierra, es decir, $$
q < Q_\oplus \quad \text{y} \quad Q > q_\oplus.
$$
   - Para valores pequeños de $ e $ (órbitas más circulares), el rango de $ a $ que permite el cruce con la órbita terrestre es más amplio.
   - A medida que  $e $ aumenta (órbitas más excéntricas), el rango de $ a $ que permite el cruce se reduce, concentrándose en la zona central del gráfico.
   - Las órbitas con valores extremos de $ a $ o $ e$ no cumplen las condiciones de cruce, ya sea porque están demasiado lejos o porque sus órbitas son demasiado excéntricas para intersectar la órbita terrestre.



```python
# Importamos las librerias necesarias
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from google.colab import files
from pandas import read_csv
import gdown
```

```python
# Definimos la distancia de la Tierra al Sol en UA
q_tierra = 0.983  # Perihelio de la Tierra en UA
Q_tierra = 1.017  # Afelio de la Tierra en UA

# Rango del semieje mayor
a_v = np.linspace(0.5, 3, 500)

# Excentricidad para las condiciones dadas
e_q = 1 - q_tierra / a_v  # Para q = Q_tierra
e_Q = Q_tierra / a_v - 1  # Para Q = q_tierra

# Crear un meshgrid más denso para la suavización de las curvas
a_grid = np.linspace(0, 3, 500)
e_grid = np.linspace(0, 1, 500)

# Generamos el meshgrid
X, Y = np.meshgrid(a_grid, e_grid)

# Creamos las curvas de excentricidad interpoladas
y_itp = np.interp(a_grid, a_v, e_q)
y2_itp = np.interp(a_grid, a_v, e_Q)

# Hacemos una máscara para delimitar la zona sombreada
mask = (Y >= y_itp) & (Y >= y2_itp)

# Aplicamos el contorno para sombrear la zona central
plt.contourf(X, Y, mask, levels=[0.5, 1], colors='pink', alpha=0.7)
# Bordes de la sombra
plt.contour(X, Y, mask, levels=[0.5, 1], colors='black', linewidths=1.5)

# Graficamos las curvas q = Q_tierra y Q = q_tierra
# Curva para q = Q_tierra
plt.plot(a_v, e_q, label=r"$q = Q_\oplus$", color='yellow', linewidth=2)
# Curva para Q = q_tierra
plt.plot(a_v, e_Q, label=r"$Q = q_\oplus$", color='green', linewidth=2)

# Etiquetas y leyenda
plt.xlabel("Semieje mayor $a$ (UA)")
plt.ylabel("Excentricidad $e$")
plt.title("Órbitas centrales que cruzan la órbita de la Tierra")
plt.grid()
plt.ylim(0, 1)
plt.legend(loc='upper right')
plt.show()
```

# Punto 3
Para demostrar que la excentricidad $ e $ de la órbita de los asteroides con inclinación orbital $i = 0 $ y constante de Tisserand $ T $ cumple la relación:

$$
e = \sqrt{1 - \frac{1}{4\bar{a}^3\cos^2 i}(T\bar{a} - 1)^2}
$$

V amos a partir de la expresión del parámetro de Tisserand $ T $ para $ i = 0 $:

$$
T = \frac{1}{\bar{a}} + 2\cos i\sqrt{\bar{a}(1 - e^2)}
$$

Tenemos $ \cos i = 1 $, por lo que la expresión se simplifica a:

$$
T = \frac{1}{\bar{a}} + 2\sqrt{\bar{a}(1 - e^2)}
$$

Ahora, vamos a despejar el término que contiene  a $e $ :

Restamos $ \frac{1}{\bar{a}} $ de ambos lados:

$$
T - \frac{1}{\bar{a}} = 2\sqrt{\bar{a}(1 - e^2)}
$$

Dividimos ambos lados por 2:

$$
\frac{T - \frac{1}{\bar{a}}}{2} = \sqrt{\bar{a}(1 - e^2)}
$$
Elevamos al cuadrado ambos lados para eliminar la raíz cuadrada:

$$
\left(\frac{T - \frac{1}{\bar{a}}}{2}\right)^2 = \bar{a}(1 - e^2)
$$

Simplificamos el lado izquierdo:

$$
\frac{(T\bar{a} - 1)^2}{4\bar{a}^2} = \bar{a}(1 - e^2)
$$

Multiplicamos ambos lados por $ 4\bar{a}^2 $:

$$
(T\bar{a} - 1)^2 = 4\bar{a}^3(1 - e^2)
$$

Dividimos ambos lados por $ 4\bar{a}^3 $:

$$
\frac{(T\bar{a} - 1)^2}{4\bar{a}^3} = 1 - e^2
$$

Despejo $ e^2 $:

$$
e^2 = 1 - \frac{(T\bar{a} - 1)^2}{4\bar{a}^3}
$$

Finalmente, tomamos la raíz cuadrada de ambos lados para obtener :

$$
e = \sqrt{1 - \frac{(T\bar{a} - 1)^2}{4\bar{a}^3}}
$$

Dado que $\cos i = 1 $ para $ i = 0 $, podemos incluir $\cos^2 i $ en el denominador para generalizar la expresión:

$$
e = \sqrt{1 - \frac{1}{4\bar{a}^3\cos^2 i}(T\bar{a} - 1)^2}
$$

# Punto 4
Al comparar las curvas definidas por la ecuación del parámetro de Tisserand $ T $ con las curvas obtenidas en el punto 2 (que representan las órbitas que cruzan la órbita de la Tierra), se pueden hacer las siguientes observaciones:

- Las curvas del parámetro de Tisserand $ T $ son curvas de nivel que representan órbitas con un valor constante de $T $. Estas curvas dependen de la relación entre $ a $,$ e $ e $ i$ (inclinación orbital), pueden extenderse a lo largo de un rango más amplio de $ a $ y $ e $.Estas tienden a tener una forma más suave y continua en comparación con las curvas del punto 2, que pueden tener una forma más restringida debido a las condiciones específicas de cruce con la órbita terrestre.

- Las curvas del punto 2 representan las condiciones $  q = Q_\oplus$ y  $Q_\oplus = q$, que definen las órbitas que cruzan la órbita de la Tierra,están más concentradas en la zona central del gráfico.

- La superposición de las curvas sugiere que los cuerpos con órbitas que tienen un valor específico de $ T $ (como $ T = 2.5, 3.0, 3.5 $) pueden ser potenciales asteroides cercanos a la Tierra (NEA) si también cumplen las condiciones de cruce.
- Las diferencias en la forma y posición de las curvas resaltan cómo el parámetro de Tisserand proporciona una perspectiva dinámica adicional para identificar y clasificar órbitas de asteroides.


```python
# Función que calcula la excentricidad según la ecuación
def calcular_e(a_barra, T):
    expr = 1 - (1 / (4 * a_barra**3)) * (T * a_barra - 1)**2
    expr[expr < 0] = np.nan  # Evita valores negativos en la raíz
    return np.sqrt(expr)

# Rango de a_barra extendido para las curvas de T
a_barra = np.linspace(0.5, 3, 400)

# Valores de T dados
T_v = [2.5, 3.0, 3.5]

# Crear el meshgrid para el sombreado
a_grid = np.linspace(0.5, 3, 500)
e_grid = np.linspace(0, 1, 500)

# Generamos el meshgrid para el sombreado
X, Y = np.meshgrid(a_grid, e_grid)

# Creamos las curvas de excentricidad interpoladas
y1_interp = np.interp(a_grid, a_v, e_q)
y2_interp = np.interp(a_grid, a_v, e_Q)

# Hacemos una máscara para delimitar la zona sombreada
mask = (Y >= y1_interp) & (Y >= y2_interp)

# Graficar
plt.figure(figsize=(5, 5))

# Graficamos las curvas para los diferentes T
for T in T_v:
    e_val = calcular_e(a_barra, T)
    plt.plot(a_barra, e_val, label=f"T = {T}", linewidth=2)

# Graficamos las curvas q = Q_tierra y Q = q_tierra
# Curva para q = Q_tierra
plt.plot(a_v, e_q, label=r"$q = Q_\oplus$", color='purple', linestyle='--', linewidth=2)
# Curva para Q = q_tierra
plt.plot(a_v, e_Q, label=r"$Q = q_\oplus$", color='purple', linestyle='--', linewidth=2)

# Aplicamos el contorno para sombrear la zona que queremos
plt.contourf(X, Y, mask, levels=[0.5, 1], colors='pink', alpha=0.7)
# Bordes de la sombra
plt.contour(X, Y, mask, levels=[0.5, 1], colors='purple', linewidths=1.5)

# Etiquetas y leyenda
plt.xlabel("Semieje mayor $a$ (UA)")
plt.ylabel("Excentricidad $e$")
plt.title("Comparación de Curvas de Excentricidad para Diferentes T ")
plt.grid()
plt.ylim(0, 1)
plt.legend(loc='upper right')
plt.show()
```
